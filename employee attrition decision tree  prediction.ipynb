```python
# importing os for input / output operation.
import os
os.chdir("D:/R and PYTHON files/csv output")# set your own working directory
os.getcwd()
# importing the necessary library package for data operation and visualization. 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# reading the data into python environment
df1=pd.read_excel("Takenmind_intern.xlsx",sheet_name='3')# left employee data set
df2=pd.read_excel("Takenmind_intern.xlsx",sheet_name='2')# existing employee data set
df3=pd.read_excel("Takenmind_intern.xlsx",sheet_name='4')# combined dataset for model preparation.
```

```python
# importing the libraries for model preparation
import sklearn
from sklearn import tree
from sklearn.metrics import accuracy_score
# library for making the data partion for train and test
from sklearn.model_selection import train_test_split
```

```python
# converting the characters to integer by assigning the levels in features like department and salary.
def score_to_numeric(x):
    if x=='sales':
        return 1
    if x=='accounting':
        return 2
    if x=='hr':
        return 3
    if x=='technical':
        return 4
    if x=='support':
        return 5
    if x=='management':
        return 6
    if x=='IT':
        return 7
    if x=='product_mng':
        return 8
    if x=='marketing':
        return 9
    if x=='RandD':
        return 10
df3['dept']=df3['dept'].apply(score_to_numeric)

def score_to_numeric(x):
    if x=='low':
        return 1
    if x=='medium':
        return 2
    if x=='high':
        return 3
df3['salary']=df3['salary'].apply(score_to_numeric)
df3['dept']=df3['dept'].astype('float')
df3['salary']=df3['salary'].astype('float')
```

```python
# data is made into two parts train and test data for model, in order to predict the new test cases. 
x=df3.values[:,0:10]
y=df3.values[:,10]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
# fitting the decision tree classifier.
clf=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
# DecisionTreeClassifier
clf.fit(xtrain,ytrain)
```

```python
# predicting  for the test cases 
predictions=clf.predict(xtest)
```

```python
# checking the accuracy of the model.
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,predictions)
cm=pd.crosstab(ytest,predictions)
```
